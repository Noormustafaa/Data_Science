```python
#List 
#List ek ordered collection hoti hai jisme:
"""""
Multiple values hoti hain

Values change ho sakti hain

Index hota hai (0 se start)

"""

marks=[34,87,20,11]
print([0])
print([2])

marks[1]=99 #mutable 
print(marks[1])
```

```python
"""
Tuple bhi list jaisi hoti hai lekin:

Values change nahi hoti

Read-only hoti hai

"""

colors = ('red', 'green', 'blue')

print(colors[0])
#colors[2]='flask' imutable
print(colors[2])

print(type(colors))
```

```python
"""Dictionary kya hoti hai?

Dictionary me data key : value ke form me hota hai."""

student = {
    "name": "Ali",
    "age": 18,
    "marks": 90
}

print(student["name"])
print(student["age"])


#Ye bhi safe tareeqa hai (agar key na ho):

print(student.get("age"))
```

```python
"""""
Step 4: Set (Unique values only)
🔹 Set kya hota hai?

Set ek collection hai:

Unique values only

No index

Unordered
"""
numbers = {1, 2, 3, 3, 4}
print(numbers)   # {1,2,3,4}
```
